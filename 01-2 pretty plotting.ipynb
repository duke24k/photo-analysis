```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
## 사진 데이터 로드 
pic_ds = pd.read_pickle('./resource/pic_ds.pkl')
pic_ds.sort('datetime', inplace=True)
```

```python
pic_ds['days'] = pic_ds['datetime'].apply(lambda x : (x - datetime.datetime(2011, 1, 1)).days)
```

```python
from datetime import timedelta
```

```python
sns.set(style="whitegrid", palette="colorblind", font_scale=1.6, rc={'font.family':'NanumGothic'} )
plt.figure(figsize=(12,6))
ax = plt.hist(pic_ds['days'], alpha=.7 ,)
plt.ylabel(u'사진수')

ticks_index, ticks_label = [], []
for day in range(430,2100,150):
    ticks_index.append(day)
    ticks_label.append((datetime.datetime(2011,1,1) + timedelta(days=day)).strftime(u'%y년 %m월'))
ticks = plt.xticks( ticks_index, ticks_label, rotation='vertical' )
```

```python
sns.set(style="whitegrid", palette="colorblind", font_scale=1.6, rc={'font.family':'NanumGothic'} )
plt.figure(figsize=(12,6))
sns.distplot(pic_ds['days'], kde=True)
plt.ylabel(u'사진수'); plt.xlabel('');
ticks_index, ticks_label = [], []
for day in range(430,2100,150):
    ticks_index.append(day)
    ticks_label.append((datetime.datetime(2011,1,1) + timedelta(days=day)).strftime(u'%y년 %m월'))
ticks = plt.xticks( ticks_index, ticks_label, rotation='vertical' )
```

```python
sns.set(style="whitegrid", palette="colorblind", font_scale=1.6, rc={'font.family':'NanumGothic'} )
plt.figure(figsize=(12,5))
sns.distplot(pic_ds['days'], kde=True)
plt.ylabel(u'사진수'); plt.xlabel('');
ticks_index, ticks_label = [], []
for day in range(430,2100,150):
    ticks_index.append(day)
    ticks_label.append((datetime.datetime(2011,1,1) + timedelta(days=day)).strftime(u'%y년 %m월'))
ticks = plt.xticks( ticks_index, ticks_label, rotation='vertical' )
x = plt.xlim(pic_ds['days'].min(), pic_ds['days'].max())
```

```python
sns.set(style="whitegrid", palette="colorblind", font_scale=1.6, rc={'font.family':'NanumGothic'} )
plt.figure(figsize=(12,6))
ax = sns.distplot(pic_ds['days'], kde=True)
plt.ylabel(u'사진수'); plt.xlabel('');
ticks_index, ticks_label = [], []
for day in range(430,2100,150):
    ticks_index.append(day)
    ticks_label.append((datetime.datetime(2011,1,1) + timedelta(days=day)).strftime(u'%y년 %m월'))
ticks = plt.xticks( ticks_index, ticks_label, rotation='vertical' )
x = plt.xlim(pic_ds['days'].min(), pic_ds['days'].max())

ax.annotate(u'Junior 탄신일(13-05)', xy=(870, .0007), xytext=(950, .0012), fontsize=20, color='r',
            arrowprops=dict(facecolor='r', shrink=0.1, width=2))
ax.annotate(u'IPhone6 구매(16-02)', xy=(1800, .0020), xytext=(1200, .00203), fontsize=20, color='r',
            arrowprops=dict(facecolor='r', shrink=0.1, width=2))
```
